# ChessBoard game 


```python
import matplotlib.pyplot as plot
import numpy as np
grid=""
grid = np.zeros((8,8,3),np.float64)
```

# Visual of black Chessboard

```python
plot.imshow(grid)
```

```python

class ChessBoard:
    def __init__(self):
      #  set all values to zero chess will be black
      self.grid = np.zeros((8,8,3))
      # set every even row and even col values to ones and those will be white
    #   [0,0]=>white  [2,2]=>white  ...  [6,6]=> white
      self.grid[0::2,0::2] = (1,1,1)
    #   [1,1]=>white [3,3]=>white  ...[7,7]=> white  
      self.grid[1::2,1::2] = (1,1,1)
      self.red=[]
      self.blue=[]

    def render(self):
        plot.imshow(self.grid)

    def add_red(self,row,col):
        self.grid[row,col]=[1,0,0]
        self.red=[row,col]

    def add_blue(self,row,col):
        self.grid[row,col]=[0,0,1]
        self.blue=[row,col]
    def is_under_attack(self):
        if self.blue[0]==self.red[0] or self.blue[1]==self.red[1] or abs(self.blue[1]-self.red[1])== abs(self.blue[0]-self.red[0]):
            print("underattack")
            return True
        else:
            print("safe zone")
            return False
```

## RUN THE CODE

```python
B1=ChessBoard()
B1.add_blue(5,2)
B1.add_red(4,4)
print(B1.is_under_attack())
B1.render()
B2=ChessBoard()
B2.add_blue(1,2)
B2.add_red(4,2)
print(B2.is_under_attack())
B2.render()
```

## Testing when Row attack

```python
# First test on rows
test_row_attack = ChessBoard()
test_row_attack.add_red(1, 0)
test_row_attack.add_blue(1, 7)
test_row_attack.render()
assert test_row_attack.is_under_attack() == True
print("-------------test_row_attack passed---------------")
```

## Testing when Coloumn attack

```python
test_col_attack = ChessBoard()
test_col_attack.add_red(3, 5)
test_col_attack.add_blue(4, 5)
test_col_attack.render()
assert test_col_attack.is_under_attack() == True
print("-------------test_col_attack passed---------------")
```

## Testing when diagonal attack

```python

test_row_diagonal = ChessBoard()
test_row_diagonal.add_red(4, 4)
test_row_diagonal.add_blue(6, 6)
test_row_diagonal.render()
assert test_row_diagonal.is_under_attack() == True
print("-------------test_row_diagonal passed---------------")

```

## Testing when no attack

```python
test_no_attack = ChessBoard()
test_no_attack.add_red(4, 4)
test_no_attack.add_blue(5, 6)
test_no_attack.render()
assert test_no_attack.is_under_attack() == False
print("-------------test_no_attack passed---------------")
```
